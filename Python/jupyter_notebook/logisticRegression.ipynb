```python
from sklearn.linear_model import LogisticRegression 
from sklearn.cross_validation import train_test_split
from sklearn import metrics
import statsmodels.api as sm
import pandas as pd
import numpy as np

df = sm.datasets.fair.load_pandas().data
df['had_affair'] = df['affairs'].apply(lambda x : 1 if x > 0 else 0)
df.head()
```

```python
occ_dummies = pd.get_dummies(df.occupation)
hus_occ_dummies = pd.get_dummies(df.occupation_husb)

occ_dummies.columns = ['occ1','occ2','occ3','occ4','occ5','occ6']
hus_occ_dummies.columns = ['hocc1','hocc2','hocc3','hocc4','hocc5','hocc6']

occ_dummies.head()

X = df.drop(['occupation', 'occupation_husb','had_affair'], axis =1) 
dummys = pd.concat([occ_dummies, hus_occ_dummies], axis =1)
X = pd.concat([X, dummys], axis=1)
X = X.drop(['occ1','hocc1','affairs'], axis = 1)

X.head()
```

```python
Y = df.had_affair
Y = np.ravel(Y)
print(Y)
log_model = LogisticRegression()
log_model.fit(X, Y)
log_model.score(X, Y)

coeff_df = pd.DataFrame([X.columns, log_model.coef_[0]]).T
coeff_df.columns = ['Variables','Coefficients']
coeff_df
```

```python
print(log_model.coef_)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)

log_model2 = LogisticRegression() 
log_model2.fit(X_train, Y_train)
class_predict = log_model2.predict(X_test)


metrics.accuracy_score(Y_test, class_predict)
```
