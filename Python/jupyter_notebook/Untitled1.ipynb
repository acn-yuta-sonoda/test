## import field

```python
import math
import numpy as np
from matplotlib import pyplot as plt
```

```python
pi = math.pi

x = np.linspace(0, 2 * pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label = 'sin')
plt.plot(x, cos_y, label = 'cos')

plt.title('sin and cos curves')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.legend()
plt.show()
```
